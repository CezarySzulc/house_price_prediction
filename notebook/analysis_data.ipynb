```python
REG_FEATURES = ['sqft_living', 'sqft_lot', 'sqft_above',
                'sqft_basement', 'yr_built', 'lat', 'long', 'sqft_living15', 'sqft_lot15']

CAT_FEATURES = ['waterfront']#, 'view', 'condition', 'grade', 'zipcode']

INT_FEATURES = ['bedrooms', 'bathrooms', 'floors', 'yr_renovated']

PRICE = 'price'

DATE = 'date'
```

```python
from xgboost import XGBRegressor
from xgboost import plot_importance

import numpy as np
```

```python
model = XGBRegressor(importance_type='total_cover')
```

```python
model.fit(df[REG_FEATURES], df[PRICE])
```

```python
a = list(df.columns)
a.remove('date')
a
```

```python
dict_imp_features = dict(zip(REG_FEATURES, model.feature_importances_))

# choose best params and create more stong features
list_sorted_importance = sorted(model.feature_importances_, reverse=True)
first_param = [col_name for col_name, imp in dict_imp_features.items() if imp == list_sorted_importance[0]]
second_param = [col_name for col_name, imp in dict_imp_features.items() if imp == list_sorted_importance[1]]

# add and multiple values form best params
str_add_column_name = '({})'.format('__+__'.join(first_param+second_param))
str_mul_column_name = '({})'.format('__*__'.join(first_param+second_param))
str_log_column_name = '(log__{})'.format(first_param[0])

if not (str_add_column_name in df.columns):
    df[str_add_column_name] = df[first_param].add(df[second_param].values).values
    df[str_mul_column_name] = df[first_param].mul(df[second_param].values).values
    REG_FEATURES.extend([str_add_column_name, str_mul_column_name])
else:
    print('{} exist in data'.format(str_add_column_name))
    print('{} exist in data'.format(str_mul_column_name))
    
if not (str_log_column_name in df.columns):
    df[str_log_column_name] = np.log(df[first_param])
    REG_FEATURES.append(str_log_column_name)
else:
    print('{} exist in data'.format(str_log_column_name))

```
